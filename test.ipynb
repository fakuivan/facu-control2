```python
%config InlineBackend.figure_formats = ['svg']
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.transforms import Affine2D
from utils.units import basis, u
from utils.mpl import nu_axes

nu = basis(unitary=False)

fig, (ax_no_units, ax_scaled, ax_no_scale) = plt.subplots(1, 3)
x = np.linspace(-4, 4, 100)*nu[u.N]
y = np.linspace(2, -2, 100)*nu[u.km/u.hour]

ax_no_units.set_title("Without units nor scaling")
ax_no_units.set_yscale("log")
ax_no_units.plot(x/nu[u.N], y/nu[u.km/u.hour])

ax_scaled.set_title("With units and scaling")
ax_scaled.set_yscale("log")
set_xlabel, set_ylabel = nu_axes(
    ax_scaled, nu, u.N, u.km/u.hour)
set_xlabel("Speed")
set_ylabel("Force")

ax_scaled.plot(x, y)

ax_no_scale.set_title("With units but no tick scaling")
ax_no_scale.set_yscale("log")
ax_no_scale.plot(x, y)

fig.set_figwidth(15)
fig.tight_layout()
```
